```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('closest.csv', index_col=0)
points = df.values
```

```python
points = points.reshape(-1, 1)
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(points)
```

```python
prediction_labels = model.predict(points)
```

```python
prediction = []
temp = []
for i in range(500000):
    if (i%500 == 0 and i !=0):
        prediction.append(temp)
        temp = [prediction_labels[i]]
    else:
        temp.append(prediction_labels[i])
prediction.append(temp)
```

```python
prediction = np.array(prediction)
```

```python
df = pd.read_csv('dropped_points.csv', index_col=0)
true = df.values
true = true.astype(int)
```

```python
def test(predictions, true_labels):
    # Predictions: an array where each row is a prediction
    # 0 for dropouts, 1 for non-dropouts

    # Return Quality_Prediction, one row per prediction
    # Giving how many true zeros are classifed and how many are not as well as
    # the average accuracy

    n, m = predictions.shape
    Quality_Prediction = np.zeros((3, n))

    for i in range(n):
        for j in range(m):
            if (predictions[i,j] == true_labels[i,j] & true_labels[i,j] == 0):
                Quality_Prediction[0, i] += 1
            elif (predictions[i,j] == true_labels[i,j] & true_labels[i,j] == 1):
                Quality_Prediction[1, i] += 1
    for i in range(n):
        for j in range(m):
            if predictions[i,j] == true_labels[i,j]:
                Quality_Prediction[2, i] += 1/m

    return Quality_Prediction
```

```python
result = test(prediction, true)
```
